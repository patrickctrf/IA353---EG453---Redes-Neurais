# Questão 4

Para treinamento da rede enural, utilizou-se o texto "Aquela Família", por ser um dos mais novos do site indicado, com vocabulário minimamente compatível com o atual (muito mais que dom casmurro, que levava a rede a produzir frases muito incomuns para nossa época) e tempo de treinamento reduzido, por ser uma obra menor.

O código abaixo faz o treinamento da rede SEM abrir o painel de iteração, deixando que o usuário faça isso no código seguinte, onde pode utilizar um modelo já salvo (muto útil para não perder os dados e poder interromper o uso do computador).

```python
from textgenrnn import textgenrnn

textgen = textgenrnn()
textgen.generate()

file_path = "aquelafamilia.txt"
textgen.reset()
textgen.train_from_file(file_path, new_model=True, num_epochs=20,gen_epochs=10, word_level=True)
textgen.generate(interactive=False, top_n=5)
textgen.save('aquelafamilia.hdf5')

```

## Código para utilização de modelos já gerados

O código abaixo permite interagir com um modelo já treinado, permitindo criar frases e até mesmo prosseguir com o treinamento (basta alterar o número de épocas).

A frase gerada ao final desta execução (um pouco antes dos erros emitidos pela interrupção do script) ficou surpreendentemente parecida com o "tom" de expressão encontrado na obra "Aquela família", onde há fortes interações de membros de uma mesma casa e de diferentes níveis hieráquicos. As opções de nome oferecidas pelo modelo são, por exemplo, "josé, maria cândida e afins, tipicamente brasileiros e clássicos da época (década de 50).

As frases geradas "francisco para dentro de si. maria joana! venha cá! ai, jesus. sabe? <'s> uma rapariga de mais! <'s><'s>" lembram significativamente até o vocabulário e forma de expressão de novelas de época, indicando um bom aprendizado e expressão em linguagem natural por parte da máquina.

```python
from textgenrnn import textgenrnn

textgen = textgenrnn(config_path="textgenrnn_config.json",weights_path="textgenrnn_weights.hdf5",vocab_path="textgenrnn_vocab.json")
textgen.train_from_largetext_file('aquelafamilia.txt',new_model=False,num_epochs=0)
textgen.save("aquelafamilia.hdf5")
textgen.generate(interactive=True, top_n=5)
```

## Temperature em 0.2

Segundo a documentação da própria biblioteca textgenrnn, o parâmetro temperature tem a função de diversificar a produção de vocabulário, sendo que quanto menor, mais parecida será a geração de frases com a do texto original estudado.

Frases geradas nesta condição: "maria celeste cala - se de novo . limpa as lágrimas ; mas sem dizer palavra"

```python
from textgenrnn import textgenrnn

textgen = textgenrnn(config_path="textgenrnn_config.json",weights_path="textgenrnn_weights.hdf5",vocab_path="textgenrnn_vocab.json")
textgen.train_from_largetext_file('aquelafamilia.txt',new_model=False,num_epochs=0)
textgen.save("aquelafamilia.hdf5")
textgen.generate(interactive=True, top_n=5, temperature=0.2)
```

## Temperature em 1.0

Com temperature mais elevado, a geração de frases se torna mais criativa e desvinculada do texto original, se tornando, porém, menos confiável e suscetível a contextos sem sentido semântico.

Frases geradas nesta condição: "maria celeste parece reanimar de novo . há um homem honrado !"

Aumentando a temperature, as frases diversificaram sem, no entanto, fugir do contexto e/ou do nexo.

```python
from textgenrnn import textgenrnn

textgen = textgenrnn(config_path="textgenrnn_config.json",weights_path="textgenrnn_weights.hdf5",vocab_path="textgenrnn_vocab.json")
textgen.train_from_largetext_file('aquelafamilia.txt',new_model=False,num_epochs=0)
textgen.save("aquelafamilia.hdf5")
textgen.generate(interactive=True, top_n=5, temperature=1.0)
```
